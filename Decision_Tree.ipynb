```python
#import library
import pandas as pd
import numpy as np

import warnings 
warnings.filterwarnings('ignore')
```

#Encoding

```python
#import library
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

#import pandas library to check the data type
from pandas.core.dtypes.common import is_numeric_dtype
```

```python
for col in x.columns:
  if is_numeric_dtype(x[col]):
    continue
  else:
    x[col]=le.fit_transform(x[col])
```

```python
x.head()
```

##1= sunny
##0= cloudy

##1= warm
##0= indoor

#Apply DecisionTreeClassifier

```python
#import library
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier()
```

```python
clf.fit(x,y)
```

```python
clf.predict([[1,0,0]])
```

```python
clf.predict([[1,0,1]])
```

```python
clf.predict([[1,1,0]])
```
